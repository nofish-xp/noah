1，        生成一个文件a.txt

2，        在文件中从100到1写入100个数字，每个文字一行

3，        关闭文件

4，        打开文件

5，        打印文件中每行行号和每行内容



```python
#1， 生成一个文件a.txt
fileName = 'a.txt'
f = open(fileName, 'w')
```

```python
#2， 在文件中从100到1写入100个数字，每个文字一行
for i in reversed(range(1, 100 + 1)):
    f.write(str(i) + "\n")
```

```python
#3， 关闭文件
f.close()
```

```python
#4， 打开文件
f = open(fileName, 'r')
```

```python
#打开文件的最好方式（练习关闭文件除外）
with open("a.txt", 'r') as fp:
    for (l,v) in enumerate(fp):
        print("行数 %d, 值 %s"%(l+1,v), end="\n")
```

```python
#5， 打印文件中每行行号和每行内容
for line in f:
    print(line, end="")
f.close()
```
